```python
class Item:
    def __init__(self, title, price, quantity):
        self._title = title
        self._price = price
        self._quantity = quantity

    @property
    def title(self):
        return self._title

    @title.setter
    def title(self, value):
        self._title = value

    @property
    def price(self):
        return self._price

    @price.setter
    def price(self, value):
        self._price = value

    @property
    def quantity(self):
        return self._quantity

    @quantity.setter
    def quantity(self, value):
        self._quantity = value
```

```python
class Artist:
    def __init__(self, name):
        self._name = name
        self._member_names = [None] * 20
        self._member_instruments = [None] * 20

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        self._name = value

    def add_member(self, member_name, instrument, index):
        if 0 <= index < 20:
            self._member_names[index] = member_name
            self._member_instruments[index] = instrument
```

```python
# from item import Item

class Book(Item):
    def __init__(self, title, price, quantity, author, publisher, category):
        super().__init__(title, price, quantity)
        self._author = author
        self._publisher = publisher
        self._category = category

    @property
    def author(self):
        return self._author

    @author.setter
    def author(self, value):
        self._author = value

    @property
    def publisher(self):
        return self._publisher

    @publisher.setter
    def publisher(self, value):
        self._publisher = value

    @property
    def category(self):
        return self._category

    @category.setter
    def category(self, value):
        self._category = value
```

```python
# from item import Item
# from artist import Artist
from datetime import datetime

class CD(Item):
    def __init__(self, title, price, quantity, artist, release_date):
        super().__init__(title, price, quantity)
        self._artist = artist
        self._release_date = release_date

    @property
    def artist(self):
        return self._artist

    @artist.setter
    def artist(self, value):
        self._artist = value

    @property
    def release_date(self):
        return self._release_date

    @release_date.setter
    def release_date(self, value):
        self._release_date = value
```

```python
# from item import Item
from datetime import datetime

class ClassicalCD(Item):
    def __init__(self, title, price, quantity, composer, recording_location, release_date):
        super().__init__(title, price, quantity)
        self._composer = composer
        self._performers = [None] * 5
        self._performer_count = 0
        self._recording_location = recording_location
        self._release_date = release_date

    @property
    def composer(self):
        return self._composer

    @composer.setter
    def composer(self, value):
        self._composer = value

    @property
    def recording_location(self):
        return self._recording_location

    @recording_location.setter
    def recording_location(self, value):
        self._recording_location = value

    @property
    def release_date(self):
        return self._release_date

    @release_date.setter
    def release_date(self, value):
        self._release_date = value

    def add_performer(self, performer):
        if self._performer_count < len(self._performers):
            self._performers[self._performer_count] = performer
            self._performer_count += 1
        else:
            print("No more room for performers!")

    def show_performers(self):
        print("Performers:")
        for i in range(self._performer_count):
            print(self._performers[i])
```
